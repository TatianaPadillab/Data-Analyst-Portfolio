```python
from pyodide.http import pyfetch

async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())
```

```python
#the URL of the data set is assigned to "filepath"
```

```python
await download(file_path, "usedcars.csv")
file_name="usedcars.csv"
```

```python
df.replace("?", np.nan, inplace = True)
df.head(20)
```

```python
#detecting missing data. "True" means the value is a missing value while "False" means the value is not a missing value.
```

```python
missing_data = df.isnull()
missing_data.head(10)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
#I'm going to replace each missing value by their mean value, specially "normalized-losses" column which presented 41 missing values
```

```python
    #---Calculating the mean value for the "normalized-losses" column
```

```python
    avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
    #---Replacing "NaN" with mean value in "normalized-losses" column
```

```python
#Calculating the mean value for the "bore" column and replacing "NaN" with the mean value in the "bore" colum
```

```python
avg_peak_rpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak-rpm:", avg_peak_rpm)
```

```python
#Using the Pandas method 'cut' to segment the 'horsepower' column into 3 bins and rearrenge them by high horsepower, medium horsepower, and little horsepower, we let the "horsepower" get its valued variable ranging instead of 48 to 288 with 59 unique values, to only 3! 
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
#Plotting the histogram of horsepower to see the distribution of horsepower.
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Finding 3 bins of equal size bandwidth by using Numpy's linspace(start_value, end_value, numbers_generated function.
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#Setting group names
group_names = ['Low', 'Medium', 'High']
```

```python
#Applying the function "cut" to determine what each value of df['horsepower'] belongs to
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Seeing the number of vehicles in each bin:
df["horsepower-binned"].value_counts()
```

```python
#Plotting the distribution of each bin:
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
